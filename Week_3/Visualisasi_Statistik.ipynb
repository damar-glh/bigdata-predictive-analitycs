# **Correlation dan Causation**

Hitung Koefisien Korelasi Pearson dan nilai-P dari 'wheel-base' dan 'price'.

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

from scipy import stats
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
```

## **Challenge**

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

path = "/content/smartphones_cleaned_v6.csv"
data = pd.read_csv(path)
data.head()
```

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

data = data.fillna(0)
bool_columns = data.select_dtypes(include='bool').columns
data[bool_columns] = data[bool_columns].astype(int)
for col in ['os', 'brand_name', 'processor_brand']:
    data[col] = pd.factorize(data[col])[0]
print(data.head())
```

```python
data
```

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

unique_labels = data["brand_name"].unique()
label_map = {label: idx for idx, label in enumerate(unique_labels)}
os = data["brand_name"].map(label_map).values
os
```

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

price_correlation = data.corr(numeric_only=True)["price"].sort_values(ascending=False)
print(f"Correlation with price:\n{price_correlation}")
```

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

from scipy import stats
df = pd.DataFrame(data)
pearson_coef, p_value = stats.pearsonr(df["internal_memory"], df["price"])
output = "The Pearson Correlation Coefficient is {} with a P-value of P = {}".format(pearson_coef, p_value)
print(output)
```

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

highest_variable = price_correlation.index[1]
highest_value = price_correlation[1]
print("{} - {}".format(highest_variable, highest_value))
```

```python
# Nama : Damar Galih
# Nim  : 22.11.5209

plt.figure(figsize=(10, 6))
sns.scatterplot(x=highest_variable, y="price", data=data, marker='o', s=100, edgecolor='w', alpha=0.8, color='red')
plt.title(f'Scatterplot of Price vs {highest_variable}\nCorrelation: {highest_value:.2f}')
plt.xlabel(highest_variable)
plt.ylabel('Price (INR)')
plt.grid(True)
fig = plt.gcf()
fig.text(0.95, 0.05, 'Damar Galih', fontsize=12, color='black',
         ha='right', va='top', alpha=0.5, rotation=0)
fig.text(0.95, 0.02, '22.11.5209', fontsize=12, color='black',
         ha='right', va='top', alpha=0.5, rotation=0)
plt.show()
```
